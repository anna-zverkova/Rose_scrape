```python
# specify which URL/web page we are going to be scraping
url = "http://rosegathering.com/alphroses.htm"

# open the url using urllib.request and put the HTML into the page variable
page = urllib.request.urlopen(url)

# parse the HTML from our URL into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, "lxml")
```

```python
# use the 'find_all' function to bring back all instances of the 'table' tag in the HTML and store in 'all_tables' variable
my_table=soup.find_all('table')[3]
# my_table
```

```python
# Creating csv

filename = "Rosegathering2.csv"

# looping through all <a> tags where we have href and saving rose name and href in the csv

with open(filename,'w',newline='',encoding='utf-8') as f:
    w = csv.writer(f)
    headers = 'Rose_name URL'
    bytes_headers = bytes(headers, 'utf-8')
    w.writerow(headers.split())

    for a in my_table.find_all('a', href=True):
        rose_name = a.text
        short_url = a['href']
        url = 'http://rosegathering.com/' + a['href']
        
        time.sleep(1)

        # open the url and put the HTML into the page variable
        req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urlopen(req).read()

        # parse the HTML from our URL into the BeautifulSoup parse tree format
        soup = BeautifulSoup(webpage, "lxml")
        my_text = []

        for p in soup.find_all('p'):
            print(p.text)
            my_text.append(p.text)
 


        w.writerow([rose_name,short_url,url,my_text])
```

```python
# skipping rows with letters 
df = pd.read_csv('Rosegathering2.csv', skiprows=26)
df.columns = ['rose_name', 'short_url', 'url', 'my_text']
df.head(5)
```

```python
type(df['my_text'][0])
```

```python
# removing rubbish from the scraped string in column my_text and putting it in new column
df['my_text_fix'] = df['my_text'].str.replace("[","")
df['my_text_fix'] = df['my_text_fix'].str.replace("]","")
df['my_text_fix'] = df['my_text_fix'].str.replace("xa0","")
df['my_text_fix'] = df['my_text_fix'].str.replace("\\","")
df['my_text_fix'] = df['my_text_fix'].str.replace("'',","")
df['my_text_fix'] = df['my_text_fix'].str.replace("''","")

# correctling those charecters to avoid encoding issues
df['my_text_fix'] = df['my_text_fix'].str.replace("é","e")
df['my_text_fix'] = df['my_text_fix'].str.replace("è","e")
df['my_text_fix'] = df['my_text_fix'].str.replace("ê","e")
df['my_text_fix'] = df['my_text_fix'].str.replace("ï","i")
df['my_text_fix'] = df['my_text_fix'].str.replace("à","a")
df['my_text_fix'] = df['my_text_fix'].str.replace("®","")
df['my_text_fix'] = df['my_text_fix'].str.replace("ç","c")
df['my_text_fix'] = df['my_text_fix'].str.replace("(TM)","")

df['rose_name'] = df['rose_name'].str.replace("é","e")
df['rose_name'] = df['rose_name'].str.replace("è","e")
df['rose_name'] = df['rose_name'].str.replace("ê","e")
df['rose_name'] = df['rose_name'].str.replace("ï","i")
df['rose_name'] = df['rose_name'].str.replace("à","a")
df['rose_name'] = df['rose_name'].str.replace("®","")
df['rose_name'] = df['rose_name'].str.replace("ç","c")
df['rose_name'] = df['rose_name'].str.replace("(TM)","")

df.head(5)
```

```python
# deleting column my_text
del df['my_text']
df.head(5)
```

```python
# spliting into colluns
df['name'], df['rest'] = df['my_text_fix'].str.split("',", 1).str
df['class'], df['rest2'] = df['rest'].str.split("nbreeding: ", 1).str
df['rose_other_text'], df['photographed'] = df['rest2'].str.split("nphotographed", 1).str
df.head(5)
```

```python
# removing no longer needed part of the string
df['name'] = df['name'].str.replace("'","")
df['class'] = df['class'].str.replace("'nclass: ","")
df['class'] = df['class'].str.replace('''"nclass: ''',"")
df['class'] = df['class'].str.replace("'class: ","")
df.head(5)
```
